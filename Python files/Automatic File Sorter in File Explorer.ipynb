# Automatic File Sorter in File Explorer

```python
#importing some important libraries for OS interaction as well as a user input function which handles raw inputs better
import os, shutil
from six.moves import input
```

```python
# Copy & Paste the folder path containing files when prompted

path = input("Enter your Folder path:") + '\\'
```

```python
files_in_dir = os.listdir(path)
```

```python
files_in_dir
```

```python
# Creating list of needed file-type folders

folder_names = []

for i in files_in_dir:
    try: 
        folder_names.append(i.split('.')[1] + ' files')
    except:
        print("skipped folders, no other files exist")
        continue
    
print("The folders you'll need to create are:", folder_names)
```

```python
## Creating folders needed

for loop in range(0,len(folder_names)):
    if not(os.path.exists(path + folder_names[loop])):
        os.makedirs(path+folder_names[loop])
        print(folder_names[loop] + " file made")
    else:
        print(folder_names[loop] + " Already exists!")
        continue
```

```python
# Used complex conditioning for finding the file name and matching to folder names and checking if the path exists as a whole
# May be worth using a dictionary to store file names, file types and folder types altogether to acces by 1 key

for file in files_in_dir:
    print(file)
    try: 
        file_type = file.split('.')[1]
        for folder in folder_names:
            if((file_type in file) and file_type in folder and not(os.path.exists(path+folder+'/'+file_type))):
                shutil.move(path+file,path+folder+'/'+file)
            else: 
                continue
    except:
        print("skipped folders")
        continue
    
```
